# Downsampling 

## Average Downsampling

```python
def downsample_avg(img,scale):
    height, width =img.shape
    
    #new size of downsampled image
    #new weight and height are floored by the scale factor 
    new_height = height // scale
    new_width = width // scale
    
    #Initializing downsampled image out of empty arrays
    downsampled_img = np.zeros((new_height, new_width), dtype=np.uint8)
    
    for i in range(0, height, scale):
        for j in range(0, width, scale):
            # Compute the mean average for the 2x2 block
            block = img[i:i+scale, j:j+scale]
            avg_value = np.mean(block)
            
            # Assign the mean value to the corresponding pixel in the downsampled image
            downsampled_img[i//scale, j//scale] = int(avg_value)
    
    return downsampled_img
    
```

```python
down_avg_1 = downsample_avg(grayscale_image,2)
down_avg_2 = downsample_avg(down_avg_1,2)

fig,ax=plt.subplots(2,2,figsize=(10,10))
ax[0,0].imshow(cv2.cvtColor(high_image, cv2.COLOR_BGR2RGB))
ax[0,0].set_title('Original Image')
ax[0,1].imshow(grayscale_image,cmap=plt.cm.gray)
ax[0,1].set_title('Grayscale Image')
ax[1,0].imshow(down_avg_1,cmap = plt.cm.gray)
ax[1,0].set_title('1st Downsample: Average')
ax[1,1].imshow(down_avg_2,cmap=plt.cm.gray)
ax[1,1].set_title('2nd Downsample: Average')
```

## Max Downsampling

```python
def downsample_max(img,scale):
    height, width =img.shape
    
    #new size of downsampled image
    new_height = height // scale
    new_width = width // scale
    
    #Initializing downsampled image out of empty arrays
    downsampled_img = np.zeros((new_height, new_width), dtype=np.uint8)
    
    for i in range(0, height, scale):
        for j in range(0, width, scale):
            # Compute the max average for the 2x2 block
            block = img[i:i+scale, j:j+scale]
            avg_value = np.max(block)
            
            # Assign the mean value to the corresponding pixel in the downsampled image
            downsampled_img[i//scale, j//scale] = int(avg_value)
    
    return downsampled_img
    
```

```python
down_max_1 = downsample_max(grayscale_image,2)
down_max_2 = downsample_max(down_max_1,2)

fig,ax=plt.subplots(2,2,figsize=(10,10))
ax[0,0].imshow(cv2.cvtColor(high_image, cv2.COLOR_BGR2RGB))
ax[0,0].set_title('Original Image')
ax[0,1].imshow(grayscale_image,cmap=plt.cm.gray)
ax[0,1].set_title('Grayscale Image')
ax[1,0].imshow(down_max_1,cmap = plt.cm.gray)
ax[1,0].set_title('1st Downsample: Max')
ax[1,1].imshow(down_max_2,cmap=plt.cm.gray)
ax[1,1].set_title('2nd Downsample: Max')
```

## Median Downsampling

```python
def downsample_median(img,scale):
    height, width =img.shape
    
    #new size of downsampled image
    new_height = height // scale
    new_width = width // scale
    
    #Initializing downsampled image out of empty arrays
    downsampled_img = np.zeros((new_height, new_width), dtype=np.uint8)
    
    for i in range(0, height, scale):
        for j in range(0, width, scale):
            # Compute the median average for the 2x2 block
            block = img[i:i+scale, j:j+scale]
            avg_value = np.median(block)
            
            # Assign the mean value to the corresponding pixel in the downsampled image
            downsampled_img[i//scale, j//scale] = int(avg_value)
    
    return downsampled_img
    
```

```python
down_median_1 = downsample_median(grayscale_image,2)
down_median_2 = downsample_median(down_median_1,2)

fig,ax=plt.subplots(2,2,figsize=(10,10))
ax[0,0].imshow(cv2.cvtColor(high_image, cv2.COLOR_BGR2RGB))
ax[0,0].set_title('Original Image')
ax[0,1].imshow(grayscale_image,cmap=plt.cm.gray)
ax[0,1].set_title('Grayscale Image')
ax[1,0].imshow(down_median_1,cmap = plt.cm.gray)
ax[1,0].set_title('1st Downsample: Median')
ax[1,1].imshow(down_median_2,cmap=plt.cm.gray)
ax[1,1].set_title('2nd Downsample: Median')
```

# Upsampling

## Nearest Neighbour Upsampling

```python
def Nearest_Neighbor(image, scale_factor):
    # Get the dimensions of the original image
    original_height, original_width = image.shape
    
    # Calculate the dimensions of the upsampled image
    new_height = original_height * scale_factor
    new_width = original_width * scale_factor

    # Initialize an empty array for the upsampled image
    upsampled_image = np.zeros((new_height, new_width), dtype=np.uint8)

    for i in range(new_height):
        for j in range(new_width):
            # Find the nearest pixel in the original image
            #To find the nearest pixel, used a flooring division
            original_i = i // scale_factor
            original_j = j // scale_factor
            
            # Assign the value of the nearest neighbor
            upsampled_image[i, j] = image[original_i, original_j]
    return upsampled_image
```

```python
upsampled_NN_1=Nearest_Neighbor(grayscale_image_low,2)
upsampled_NN_2=Nearest_Neighbor(upsampled_NN_1,2)

fig,ax=plt.subplots(2,2,figsize=(10,10))
ax[0,0].imshow(cv2.cvtColor(low_image, cv2.COLOR_BGR2RGB))
ax[0,0].set_title('Original Image')
ax[0,1].imshow(grayscale_image_low,cmap=plt.cm.gray)
ax[0,1].set_title('Grayscale Image')
ax[1,0].imshow(upsampled_NN_1,cmap = plt.cm.gray)
ax[1,0].set_title('1st Upsample: Nearest Neighbour')
ax[1,1].imshow(upsampled_NN_2,cmap=plt.cm.gray)
ax[1,1].set_title('2nd Upsample: Nearest Neighbour')
```

## Bilinear Upsampling

```python
def Bilinear(img, scale_factor):
    #Get dimensions of the original image
    original_height, original_width = img.shape
    #Calculate dimensions of the new image
    new_height = (original_height  * scale_factor )
    new_width = (original_width  * scale_factor )

    #Initialize new image with zeros
    new_img = [[0 for _ in range(new_width)] for _ in range(new_height)]
    #Initialize dummy image for storing the original values as reference for interpolation
    dummy = [[0 for _ in range(new_width)] for _ in range(new_height)]
    #Initialize list for calculating the mean of evaluated pixels
    appended_value = []
    
    # Copy original pixels to their new positions
    for y in range(original_height):
        for x in range(original_width):
            new_img[y * scale_factor][x * scale_factor] = img[y][x]
            dummy[y * scale_factor][x * scale_factor] = img[y][x]
            
    for y in range (new_height):
        for x in range (new_width):
            appended_value = []
            #Accessing all first neighbours
            for neighbor_y in range (-1,2):
                for neighbor_x in range (-1,2):
                    #Bound Constraints
                    if ( new_width>(x+neighbor_x)>=0 and new_height>(y+neighbor_y)>=0):
                        #Value would only be appended if the accessed neighbours is non-zero
                        if(dummy[y+neighbor_y][x+neighbor_x]!=0):
                            appended_value.append(dummy[y+neighbor_y][x+neighbor_x])
            #Calculating mean 
            mean_value = np.mean(appended_value)
            
            #Precautions if there is no appended value due to the accessed neighbours are all zeroes
            if np.isnan(mean_value):
                new_img[y][x] = 0
            else:
                new_img[y][x] = int(mean_value)

    return new_img
```

```python
upsampled_Bilinear_1=np.array(Bilinear(grayscale_image_low,2))
upsampled_Bilinear_2=np.array(Bilinear(upsampled_Bilinear_1,2))

fig,ax=plt.subplots(2,2,figsize=(10,10))
ax[0,0].imshow(cv2.cvtColor(low_image, cv2.COLOR_BGR2RGB))
ax[0,0].set_title('Original Image')
ax[0,1].imshow(grayscale_image_low,cmap=plt.cm.gray)
ax[0,1].set_title('Grayscale Image')
ax[1,0].imshow(upsampled_Bilinear_1,cmap = plt.cm.gray)
ax[1,0].set_title('1st Upsample: Bilinear')
ax[1,1].imshow(upsampled_Bilinear_2,cmap=plt.cm.gray)
ax[1,1].set_title('2nd Upsample: Bilinear')
```

## Bicubic Upsampling

```python
#The process of accessing neighbours on bicubic and bilinear is similar
#Its difference would be the range it has and difference in significance between first neighbour and second neighbour
def Bicubic(img, scale_factor):
    # Get dimensions of the original image
    original_height, original_width = len(img), len(img[0])
    # Calculate dimensions of the new image
    new_height = (original_height  * scale_factor )
    new_width = (original_width  * scale_factor )

    # Initialize new image with zeros
    new_img = [[0 for _ in range(new_width)] for _ in range(new_height)]
    dummy = [[0 for _ in range(new_width)] for _ in range(new_height)]
    appended_value = []
    # Copy original pixels to their new positions
    for y in range(original_height):
        for x in range(original_width):
            new_img[y * scale_factor][x * scale_factor] = img[y][x]
            dummy[y * scale_factor][x * scale_factor] = img[y][x]
            
    for y in range (new_height):
        for x in range (new_width):
            appended_value=[]
            for neighbor_y in range (-2,3):
                for neighbor_x in range (-2,3):
                    if ( new_width>(x+neighbor_x)>=0 and new_height>(y+neighbor_y)>=0):
                        if(dummy[y+neighbor_y][x+neighbor_x]!=0):
                            
                            
                            #Taking the first neighbour significance as 0.75 and second neighbour as 0.25
                            if(-1<=(x+neighbor_x)<2 and -1<=(y+neighbor_y)<2):
                                appended_value.append(dummy[y+neighbor_y][x+neighbor_x]*0.75)
                            else:
                                appended_value.append(dummy[y+neighbor_y][x+neighbor_x]*0.25)
                                
                                
                            
            mean_value = np.mean(appended_value)

            if (np.isnan(mean_value)):
                mean_value = 0
            new_img[y][x] = (int(mean_value))
    return new_img
```

```python
upsampled_Bicubic_1=np.array(Bicubic(grayscale_image_low,2))
upsampled_Bicubic_2=np.array(Bicubic(upsampled_Bicubic_1,2))

fig,ax=plt.subplots(2,2,figsize=(10,10))
ax[0,0].imshow(cv2.cvtColor(low_image, cv2.COLOR_BGR2RGB))
ax[0,0].set_title('Original Image')
ax[0,1].imshow(grayscale_image_low,cmap=plt.cm.gray)
ax[0,1].set_title('Grayscale Image')
ax[1,0].imshow(upsampled_Bicubic_1,cmap = plt.cm.gray)
ax[1,0].set_title('1st Upsample: Bicubic')
ax[1,1].imshow(upsampled_Bicubic_2,cmap=plt.cm.gray)
ax[1,1].set_title('2nd Upsample: Bicubic')
```

# Analysis 
## Downsampling
Downsampling images methods have differences from one method to another. Maximum valued-based downsampling are the one with the 'brightest' images as it takes the maximum value of a pixel inside a block of pixels. Meanwhile, mean and median based downsampling are similar to one another. This situation is perhaps the image don't have big variance, making the distribution of pixel values close to a normal condition.
<br>
## Upsampling 
Upsampling images on the other hand, showed more difference for each method. Each methods are seemed to be improved from one another. Nearest neighbour upsampling seemed to have the lowest performance. On the other hand, bilinear seemed to have a better performance than nearest neighbour but still inferior to bicubic operations. This would likely due to the complexity of pixel sampling process. Nearest neighbour method only takes the closest pixel from its position, bilinear could take the average of all of its first neighbours but only on a uniform scale, while bicubic can adjust the scale of the sampled pixels. In this case, the bicubic would take 0.75 significance level of the first neighbour and 0.25 of the second neighbour, while on the other hand bilinear only take the mean of all nonzeroes first neighbours.
